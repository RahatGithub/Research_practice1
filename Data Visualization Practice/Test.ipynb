```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
# plt.show()
```

```python
x = np.linspace(0,10,1000)
y = np.cos(x)
plt.plot(x,y, color='red')
plt.title('Cosine function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color='red')
plt.plot(x, np.cos(x), color='orange', linestyle='dotted')
plt.xlabel('x')
plt.show()
```
